# `Backword Elimination`

Backward elimination is a feature selection technique used in machine learning to identify a subset of the most relevant features from your data.

```python
# Backward elimination starts with all features included in the model and then removes the least relevant features one at a time.
import pandas as pd
from mlxtend.feature_selection import SequentialFeatureSelector

dataset = pd.read_csv('https://raw.githubusercontent.com/AqueeqAzam/DS-and-ML-datasets/main/loan.csv')
dataset.head()

# x = dataset.iloc[:, :-1] uses indexing to extract a specific portion of a DataFrame (dataset) stored in the variable dataset.
# [:, :] would select all rows (: for all rows) and all columns (: for all columns).
# :-1 is a special slicing notation used in Python to select all elements up to, but not including, the last element.

x = dataset.iloc[:, :-1]
y = dataset['default']

from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

# k_features=5: This argument specifies the number of features you want to select.
# forward=True indicates a forward selection approach.
# cv=3: This argument indicates the use of cross-validation with 3 folds
fs = SequentialFeatureSelector(lr, k_features=5, forward=False)
fs.fit(x, y)

# display feature names
fs.feature_names

# check score
fs.k_score_


```

# `Simple Linear Regression`

relationship between one independent variable and one dependent variable

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('https://raw.githubusercontent.com/AqueeqAzam/data-science-and-machine-learning-datasets/main/customer.csv')
data.head(5)

# take one input value of dependent variable
x = data[['Annual Income (k$)']]

# check dimension
x.ndim

y = data[['Spending Score (1-100)']]

# model training
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# selection algo
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(x_train, y_train)

# make prediction for new data point
pv = lr.predict([[20]])

# print predicted value
print('Predicted value is:', pv)

# check score
lr.score(x_test, y_test)

# check coffecient
lr.coef_

# check intercept
lr.intercept_

# draw straight line
y_pred = lr.predict(x)
sns.scatterplot(x = 'Annual Income (k$)', y = 'Spending Score (1-100)', data=data)
plt.plot(data['Annual Income (k$)'], y_pred, c='red')
plt.legend(['org', 'predict line'])
plt.show()
```

# `Confusion Matrix`

A confusion matrix is a table that is used to define the performance of a classification algorithm.

Actual data -> Confusion matrix -> Predicted data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("https://raw.githubusercontent.com/AqueeqAzam/data-science-and-machine-learning-datasets/main/students-placement.csv")
df.head(4)

x = df.drop(columns='placed', axis=1)
y = df['placed']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()

lr.fit(x_train, y_train)

y_pred = lr.predict(x_test)

from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score
cm = confusion_matrix(y_test, y_pred)
cm

sns.heatmap(cm, annot=True)
plt.show()

precision_score(y_test, y_pred)
recall_score(y_test, y_pred)
f1_score(y_test, y_pred)*100
```

# `Imbalanced Dataset`



Reduce majority of the class so that it will hace same no of as the minority

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/AqueeqAzam/data-science-and-machine-learning-datasets/main/online-ads.csv')
df.head(4)

df['Purchased'].value_counts()

# pd.get_dummies is a function from the pandas library used for one-hot encoding of categorical variables.
# columns=['Gender'] specifies that we only want to encode the column named "Gender".
df = pd.get_dummies(df, columns=['Gender'], drop_first=True)  # drop_first to avoid multicollinearity

x = df.drop(columns='Purchased', axis=1)
y = df['Purchased']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()

lr.fit(x_train, y_train)

# make prediction
y_pred = lr.predict(x_test)
y_pred

from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score
cm = confusion_matrix(y_test, y_pred)
cm

# check score
precision_score(y_test, y_pred)
recall_score(y_test, y_pred)
f1_score(y_test, y_pred)*100

# display graph
sns.heatmap(cm, annot=True)
plt.show()
```

`UnderSampleing`

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/AqueeqAzam/data-science-and-machine-learning-datasets/main/online-ads.csv')
df.head(4)

df['Purchased'].value_counts()

df = pd.get_dummies(df, columns=['Gender'], drop_first=True)

x = df.drop(columns='Purchased', axis=1)
y = df['Purchased']

from imblearn.under_sampling import RandomUnderSampler
rus = RandomUnderSampler(random_state=42)
x_rus, y_rus = rus.fit_resample(x, y)
y_rus.value_counts()

x_rus, y_rus

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_rus, y_rus, test_size=0.2, random_state=42)

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()

lr.fit(x_train, y_train)

# make prediction
y_pred = lr.predict(x_test)
y_pred

score = lr.score(x_test, y_test)
score
```

# `Naive Bayes`

Is is a classfication algo besed on bayes theorem, that describe the probebility of an event.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions

df = pd.read_csv("https://raw.githubusercontent.com/AqueeqAzam/DS-and-ML-datasets/main/credit.csv")
df.head(4)

df.isnull().sum()

sns.scatterplot(x = 'Age', y = 'Salary', data=df, hue='Purchased')
plt.show()

x = df.iloc[:, :-1]
y = df['Purchased']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
gnb = GaussianNB() # Create an instance of the GaussianNB class
gnb.fit(x_train, y_train)
gnb.score(x_test, y_test)*100
gnb.score(x_train, y_train)*100

mnb = MultinomialNB() # Create an instance of the GaussianNB class
mnb.fit(x_train, y_train)
mnb.score(x_test, y_test)*100
mnb.score(x_train, y_train)*100

bnb = BernoulliNB() # Create an instance of the GaussianNB class
bnb.fit(x_train, y_train)
bnb.score(x_test, y_test)*100
bnb.score(x_train, y_train)*100

# make prediction
bnb.predict([[34, 26000]])
```

# `Decision Tree`

IT is used for both classification and refression, but use for classification.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions

df = pd.read_csv("https://raw.githubusercontent.com/AqueeqAzam/DS-and-ML-datasets/main/credit.csv")
df.head(4)

df.isnull().sum()

x = df.iloc[:, :-1]
y = df['Purchased']

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(x)
x = pd.DataFrame(sc.transform(x), columns = x.columns)
x

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(criterion='entropy')

sns.scatterplot(x='Age', y='Salary', data=df, hue='Purchased')
plt.show()

dtc.fit(x_train, y_train)
dtc.score(x_test, y_test)*100

dtc.predict([[10, 30000]])

from sklearn.tree import plot_tree
plot_tree(dtc)
plt.show()

```

# `Unsupervised Machine Learning`

Used for unlabelled data means does not have any pre-existing labels

Generally use for categorised

Input row data (cat and dog) -> Interpretation -> Algo -> Processing -> Output -> cat % dog

Clustring: Making group

k-Means: It is clustering is an unsupervised leanring algo, whcih groups the unlabelled dataset into different clusters.

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline

df = pd.read_csv("https://raw.githubusercontent.com/AqueeqAzam/data-science-and-machine-learning-datasets/main/Employee.csv")
df

plt.scatter(df.Age, df['Income'])

km = KMeans(n_clusters=2)
y_pred = km.fit_predict(df[['Age', 'Income']])
y_pred

# make with dataframe
df['cluster'] = y_pred
df.head()

km.cluster_centers_

# visualization
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]

# Corrected plotting code
plt.scatter(df1.Age, df1['Income'], color='green') # Use comma to separate x and y data
plt.scatter(df2.Age, df2['Income'], color='red')
plt.scatter(df3.Age, df3['Income'], color='black')

plt.xlabel('Age')
plt.ylabel('Income')
plt.legend()


scaler = MinMaxScaler()
scaler.fit(df[['Income']])
df['Income'] = scaler.transform(df[['Income']])
df.head()


# elbo plot

sse = []
k_rng = range(2, 10)

for k in k_rng:
  km = KMeans(n_clusters=k)
  km.fit(df[['Age', 'Income']])
  sse.append(km.inertia_)

plt.xlabel('k')
plt.ylabel('Sum of square error')
plt.plot(k_rng, sse)
```

# `Association`

It is technique used to dicover intersting relationship or patterns in large datasets.

customer1 -> milk, bread, butter

customer2 -> milk, cake, bread, butter

customer3 -> suger, egg, milk

Application -> Market analysis, medical diagnosis, protein sequence

```python
ds = [['milk', 'onion', 'beans','eggs', 'cakes'],
      ['horlicks', 'onion', 'butter','eggs', 'chocolates'],
      ['milk', 'apple', 'beans','eggs'],
      ['milk', 'unicorn', 'beans','butter'],
      ['milk', 'onion', 'beans','ice cream', 'eggs']]

import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

ds

te = TransactionEncoder()
y_label = te.fit(ds).transform(ds)

df = pd.DataFrame(y_label, columns=te.columns_)
df

from mlxtend.frequent_patterns import apriori

fre_itemsets = apriori(df, min_support=0.5, use_colnames=True)
fre_itemsets

# Association rules mining
from mlxtend.frequent_patterns import association_rules
res = association_rules(fre_itemsets, metric="confidence", min_threshold=0.6)
res
```
